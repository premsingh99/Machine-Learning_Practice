### One-Hot-Encoding with many categories

```python
df = pd.read_csv('mercedes.csv')
```

```python
df.head()
```

```python
df= df[['X0','X1','X2','X3','X4','X5','X6']]
```

```python
df.head()
```

```python
for i in df.columns:
    print(i,len(df[i].unique()))
```

```python
df.X1.value_counts().sort_values(ascending=False)
```

```python
df.X1.value_counts().sort_values(ascending=False).head(10)
```

```python
first_10 = df.X1.value_counts().sort_values(ascending=False).head(10).index
first_10 
```

```python
first_10 = list(first_10)
first_10
```

```python
import numpy as np
for category in first_10:
    df[category] = np.where(df['X1']==category,1,0)
```

```python
df
```

```python
first_10.append('X1')
```

```python
df[first_10].head(20)
```
