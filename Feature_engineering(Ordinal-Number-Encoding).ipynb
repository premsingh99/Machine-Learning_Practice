### Ordinal number Encoding

```python
import datetime
```

```python
today_date = datetime.datetime.today()
```

```python
today_date
```

```python
##### List comprehension
days =  [today_date-datetime.timedelta(x) for x in range(15)]
```

```python
import pandas as pd
data = pd.DataFrame(days)
data.columns=['Days']
```

```python
data.head()
```

```python
data['Weekday']=data['Days'].dt.weekday_name
```

```python
data['Weekday']
```

```python
dictionary={
    'Monday':1,'Tuesday':2, 'Wednesday':3,'Thursday':4,'Friday':5,'Saturday':6,'Sunday':7
}
```

```python
dictionary
```

```python
data['Weekday_ordinal']=data['Weekday'].map(dictionary)
```

####  Count OR Frequency Encoding

```python
train_set = pd.read_csv('http://archive.ics.uci.edu/ml/machine-learning-databases/adult/adult.data',header=None)
```

```python
train_set
```

```python
train_set[1].unique()
```

```python
## using columns with categorical features only
columns = [1,3,5,6,7,8,9,13]
```

```python
train_set = train_set[columns]
train_set
```

```python
train_set.columns = ['Employment', 'Degree', 'Marital_status', 'Designation', 'Family_status', 'Race', 'Sex', 'Country' ]
```

```python
train_set
```

```python
for column in train_set.columns:
    print(column, ':', len(train_set[column].unique()))
```

```python
train_set['Country'].value_counts().to_dict()
```

```python
Country_map = train_set['Country'].value_counts().to_dict()
```

```python
train_set['Country'] = train_set['Country'].map(Country_map)
train_set.head(20)
```

#### Advantages
1. Easy to implement
2. not increasing feature space


#### Disadvantages
1. It will provide same weigh if frequency of categories are same


### Target Guided Ordinal Encoding
1. Ordering the labels according to target
2. Replace the labels by the joint probability of being 1 or 0

kaggle Dataset - Titanic - Machine Learning From Disaster

https://www.kaggle.com/c/titanic/data

```python
df = pd.read_csv('titanic_train.csv')
```

```python
df
```

```python
df = df[['Cabin', 'Survived']]
df
```

```python
df['Cabin'].fillna('Missing',inplace=True)
```

```python
df
```

```python
df['Cabin'].astype(str).str[0]
```

```python
df['Cabin']=df['Cabin'].astype(str).str[0]
df['Cabin'].unique()
```

```python
df.groupby(['Cabin'])['Survived'].mean()
```

```python
df.groupby(['Cabin'])['Survived'].mean().sort_values()
```

```python
df.groupby(['Cabin'])['Survived'].mean().sort_values().index
```

```python
ordinal_labels = df.groupby(['Cabin'])['Survived'].mean().sort_values().index
ordinal_labels
```

```python
{k:i for i,k in enumerate(ordinal_labels)}
```

```python
ordinal_labels2 = {k:i for i,k in enumerate(ordinal_labels)}
```

```python
df['Cabin_ordinal_labels']=df['Cabin'].map(ordinal_labels2)
df.head()
```

### Mean Encoding

```python
df.groupby(['Cabin'])['Survived'].mean().to_dict()
```

```python
mean_ordinal = df.groupby(['Cabin'])['Survived'].mean().to_dict()
```

```python
df['Mean_ordinal_encode'] = df['Cabin'].map(mean_ordinal)
```

```python
df
```

#### Advantages
1. captures information within the label thus rendering more predictive features
2. creates monotonic relationship between variable and the target
